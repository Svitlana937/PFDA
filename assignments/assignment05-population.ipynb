```python
# Importing the pandas library.
import pandas as pd
```

```python
url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FY006A/CSV/1.0/en"
df = pd.read_csv(url)
df.tail(3)
```

```python
# Cleaning the "Single Year of Age" and "VALUE" columns.
df["Single Year of Age"] = df["Single Year of Age"].str.replace("Under 1 year", "0")
df["Single Year of Age"] = df["Single Year of Age"].str.replace("\\D", "", regex=True)
df["Single Year of Age"] = pd.to_numeric(df["Single Year of Age"], errors="coerce")
df["VALUE"] = pd.to_numeric(df["VALUE"], errors="coerce")
```

```python
# Calculating the mean age for males.
df_male = df[df["Sex"] == "Male"]
total_age_male = (df_male["Single Year of Age"] * df_male["VALUE"]).sum()
total_pop_male = df_male["VALUE"].sum()
mean_age_male = total_age_male / total_pop_male
print(mean_age_male)
```

```python
# Calculating the mean age for females.
df_female = df[df["Sex"] == "Female"]
total_age_female = (df_female["Single Year of Age"] * df_female["VALUE"]).sum()
total_pop_female = df_female["VALUE"].sum()
mean_age_female = total_age_female / total_pop_female
print(mean_age_female)
```

```python
# Calculating the difference in mean age between females and males.
difference = mean_age_female - mean_age_male
print(difference)
```
