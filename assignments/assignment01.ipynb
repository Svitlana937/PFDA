```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading CSV data
url = 'https://raw.githubusercontent.com/andrewbeattycourseware/PFDA-courseware/main/code/data/projectedbirths-cso.csv'
data_table = pd.read_csv(url)
```

```python
# Making a plot
# plt.figure(figsize=(10,5))
plt.plot(data_table['Year'], data_table['VALUE'], 'o-', color='green')
plt.title('Forecast of births')
plt.xlabel('Year')
plt.ylabel('Count of births')
plt.grid(True)
plt.tight_layout()
plt.show()
```
